```python
import pandas as pd
import numpy as np
import os
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import tensorflow as tf
from PIL import Image
```

```python
# Define the folder path
folder_path_covid = "Covid-19"

# Create an empty list to store image paths
image_paths_covid = []
images_covid=[]

# Use os.walk to iterate through the folder and its subfolders
for root, directories, files in os.walk(folder_path_covid):
    for filename in files:
        # Check if the file is an image
        if filename.endswith(".jpg") or filename.endswith(".png") or filename.endswith(".jpeg"):
            # Build the image path and append it to the list
            image_path = os.path.join(root, filename)
            image_paths_covid.append(image_path)
            image = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2GRAY)
            images_covid.append(image)

#Create a list of labels (all "covid" in this case)
labels = ["covid"] * len(image_paths_covid)

#Create a Pandas DataFrame
df_covid = pd.DataFrame(data={
    "image": images_covid,
    "label": labels,
})
images_covid = np.array(images_covid)
# Print the DataFrame
print(" # of Covid Images : " , len(df_covid))
print("Shape of each image : ",df_covid.iloc[0,0].shape)
```

```python
# Define the folder path
folder_path_normal = "Normal"

# Create an empty list to store image paths
image_paths_normal = []
images_normal=[]
# Use os.walk to iterate through the folder and its subfolders
for root, directories, files in os.walk(folder_path_normal):
    for filename in files:
        # Check if the file is an image
        if filename.endswith(".jpg") or filename.endswith(".png") or filename.endswith(".jpeg"):
            # Build the image path and append it to the list
            image_path = os.path.join(root, filename)
            image_paths_normal.append(image_path)
            image = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2GRAY)
            images_normal.append(image)
# Create a list of labels (all "covid" in this case)
labels = ["normal"] * len(image_paths_normal)

# Create a Pandas DataFrame
df_normal = pd.DataFrame(data={
    "image": images_normal,
    "label": labels,
})

# Print the DataFrame
print(" # of Normal Images : " , len(df_normal))
print("Shape of each image : ",df_normal.iloc[0,0].shape)
```

```python
# Combine dataframes
train_df = pd.concat([df_covid.iloc[:-20], df_normal.iloc[:-20]], ignore_index=True)
print("Train dataframe:",train_df.head())
print("...")
print("# of Training Samples : ", len(train_df))
```

```python
# Combine dataframes
test_df = pd.concat([df_covid.iloc[-20:], df_normal.iloc[-20:]], ignore_index=True)
print("Test dataframe:",test_df)
print("...")
print("# of Testing Samples : ", len(test_df))
```

```python
# Display the first three images in df_covid
df_covid_sample = df_covid.head(3)
df_normal_sample = df_normal.head(3)
plt.figure(figsize=(10, 4))

for i in range(3):
    # Convert the string representation of the image to a NumPy array
    image_array = df_covid_sample['image'].iloc[i]
    
    # Create a subplot for each image
    plt.subplot(2, 3, i+1)
    
    # Display the image
    plt.imshow(image_array)
    
    # Set the title as the corresponding label
    plt.title(df_covid_sample['label'].iloc[i])

for i in range(3):
    # Convert the string representation of the image to a NumPy array
    image_array = df_normal_sample['image'].iloc[i]
    
    # Create a subplot for each image
    plt.subplot(2, 3, i+4)
    
    # Display the image
    plt.imshow(image_array)
    
    # Set the title as the corresponding label
    plt.title(df_normal_sample['label'].iloc[i])

# Adjust layout for better visualization
plt.tight_layout()
plt.colorbar()
# Show the plot
plt.show()
```

```python
train_df["image"]=train_df["image"]/255.0
test_df["image"]=test_df["image"]/255.0
```

```python
train_images=np.array([arr for arr in train_df.iloc[:,0]])
print(train_images.dtype)
test_images=np.array([arr for arr in test_df.iloc[:,0]])
print(test_images.dtype)

train_labels = []
for lbl in train_df.iloc[:,1]:
    if lbl=="covid":
        train_labels.append(1)
    else:
        train_labels.append(0)
test_labels = []
for lbl in test_df.iloc[:,1]:
    if lbl=="covid":
        test_labels.append(1)
    else:
        test_labels.append(0)

train_labels=np.array(train_labels)
test_labels=np.array(test_labels)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(128, 128)),
    tf.keras.layers.Dense(400, activation='relu'),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(2)
])
model.summary()
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model.fit(train_images, train_labels, epochs=20)
```

```python
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)

print('\nTest accuracy:', test_acc)
```

```python
probability_model = tf.keras.Sequential([model, 
                                         tf.keras.layers.Softmax()])
```

```python
predictions = probability_model.predict(test_images)
```

```python
print(predictions)
```

```python
for i in range(len(predictions)):
    if predictions[i][0]<=predictions[i][1]:
        print(f"Sample {i+1} is having Covid\n")
    else:
        print(f"Sample {i+1} is normal\n")
```
